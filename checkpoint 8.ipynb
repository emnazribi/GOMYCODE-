# hierarchical clustering 

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(df)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

# standardization and normalization

```python
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
import pandas as pd
scaler=StandardScaler()
darray=scaler.fit_transform(df)
d = preprocessing.normalize(darray, axis=0)
df1=pd.DataFrame(d)
df1.columns=df.columns

df1.head()




```

# k-means clustering

k=5

```python
df1=df1[['PURCHASES','BALANCE']]
```

```python
df1
```

```python

from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=3, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df1)
```

```python
#kmeans.predict(df1)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
labels=pd.DataFrame(kmeans.labels_)
df1['labels']=labels
```

```python
df1
```

# Elbow method

```python
sum_of_squared_distance=[]
k=range(1,17)
for i in k :
    km=KMeans(n_clusters=i)
    km.fit(df1[["PURCHASES","BALANCE"]])
    sum_of_squared_distance.append(km.inertia_)
```

```python
df1['labels'].value_counts()
```

```python
plt.plot(k,sum_of_squared_distance,'bx-')
plt.xlabel('k')
plt.ylabel('sum_of_squared_distance')
plt.title('elbow methods')
plt.show()
```
