# Question 1

```python
def F(Y,t):
    return [Y[1], -math.sin(Y[0]) - Y[1] / 4]
```

```python
def pendule(v,tmax):
    T = np.linspace(0,60,301)
    Z = si.odeint(F,[0,v],T)
    X = [z[0] for z in Z]
    V = [z[1] for z in Z]
    return X, V, T
```

# Question 2

```python
 
def diagramme_phase(V,tmax):
    for v in V:
        X,V,T= pendule(v,tmax)
        plt.plot(X,V,label ="v={}".format(v))
    plt.xlabel("position")
    plt.ylabel("vitesse")
    plt.grid()
    plt.title("Diagramme de phase")
    plt.legend()
    plt.show()
```

```python
diagramme_phase([2,4],30)
```

```python

def vitesse(k,err):
    v=1
    X,V,T= pendule(v,100)
    while X[-1] < (2 * k -1) *math.pi:
        v += 1
        X,V,T= pendule(v,100)
    vmax = v
    vmin = v-1
    while vmax - vmin > err:
        v = (vmax + vmin) / 2
        X,V,T= pendule(v,100)
        if X[-1] < (2 * k -1) *math.pi:
            vmin = v
        else:
            vmax = v
    return vmax
for k in range(10):
    print("Pour faire ",k," tours: v=",vitesse(k,1e-10))
    
```
